```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #sklearn - shortcut for Scikit-learn
# linear_model - contains algorithms related to linear models - Ridge , Lasso , logistic regression
#LinearRegression - Predicts continuous values , fit straight line to data , model relationship between input and output

```

```python
#random - module inside numpy to generate random numbers 
#seed() - function to initialize random number generator and (42) it will give same data whenever we run this file
np.random.seed(42)

# here rand generates random numbers between 0 to 1 with equal probability and (50,1) is the shape , 50 rows and 1 column 
#All generated values are between 0 to 1 , so to extent them we multiply them with 100, so it will be between 0 to 100
#rand - random numbers from uniform distribution ( range = 0 to 1) , (flat distribution)
X = np.random.rand(50,1)*100

#3.5 is slope here and X is input , the real equation - Y = mx+b , so here m is slope (3.5) and b is intercept
#randn - random numbers from normal distribution (range = -infinity to +infinity) , mean = 0 , std = 1 , (bell curve distribution)
# to add noise we are using randn so we can have negative values also
# by multiplying 20 we are increasing the randomness
# (Y = slope * X + noise)
Y = 3.5 * X + np.random.randn(50,1)*20
```

```python
model = LinearRegression()

#fit() - it trains model using given data
#here it tries different values of slope and intercept and then calculates error 
#for which values its getting the low error , that wil be the best values for slope and intercept
model.fit(X , Y)
```

```python
#predict() - method used to generate predictions
# giving same training data as input , so it will predict outputs from the trained data
Y_pred = model.predict(X)

#Y - contains random noise but Y_pred don't
#Y - actual data (scattered) example - X =10 , Y =47
#Y_pred - straight line predcited - X = 10 , Y_red = 36 
#here the model predicts 36 but the actual value was 47 so difference between this is error
```

```python
# create plotting canvas , 8 - width and 6 - height
plt.figure(figsize = (8,6))

#scatter - creates scatter plot , X - horizontal , Y - vertical(noise data)
plt.scatter(X,Y,color='blue',label='Data points')

#plt.plot() - used to draw lines , Y_pred - straight line equation , linewidth - thickness of line
plt.plot(X , Y_pred , color='red',linewidth=2,label="Regression Line")
plt.title('Linear regression on random dataset')
plt.xlabel('X')
plt.ylabel('Y')

#This shows legent box at top left corner in graph
plt.legend()

#add grid lines in background
plt.grid(True)
plt.show()
```

```python
# after running model.fit - model calculates best slope and it get stored into model.coef
#[0][0] - the value get stored in 2d so to just get the value we use this , first[0] - gives first row , second[0] - gives first element inside that row
print("Slope (coefficient):" , model.coef_[0][0])

#after running model.fir - model calculates intercept
print("Intercept:",model.intercept_[0])
```
