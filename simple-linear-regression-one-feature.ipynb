```python
import pandas as pd
# You can find this dataset from kaggle - https://www.kaggle.com/datasets/karthickveerakumar/salary-data-simple-linear-regression

df = pd.read_csv("/kaggle/input/salary-data-simple-linear-regression/Salary_Data.csv")

print(df.head())
```

```python
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
X = df[['YearsExperience']]
Y = df['Salary']
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test ,Y_train,Y_test = train_test_split (
    X , Y , test_size=0.2 , random_state=42
) 
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

#fit() - it trains model using given data
#here it tries different values of slope and intercept and then calculates error 
#for which values its getting the low error , that wil be the best values for slope and intercept
model.fit(X_train , Y_train)
```

```python
#predict() - method used to generate predictions
# giving same training data as input , so it will predict outputs from the trained data


#Y - contains random noise but Y_pred don't
#Y - actual data (scattered) example - X =10 , Y =47
#Y_pred - straight line predcited - X = 10 , Y_red = 36 
#here the model predicts 36 but the actual value was 47 so difference between this is error

Y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
MSE = mean_squared_error(Y_test , Y_predict)
print("MSE: ",MSE)
```

```python
r2 = r2_score(Y_test,Y_predict)
print("r2:",r2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize = (8,6))
plt.scatter(X_train,Y_train,color='blue',label='data points')
plt.plot(X_test,Y_predict,color='red',linewidth=2,label='regression line')
```

```python
plt.scatter(X_train,Y_train,color='blue',label='data points')
plt.plot(X_test,Y_predict,color='red',label='regression line')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("linear regression plot")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# after running model.fit - model calculates best slope and it get stored into model.coef
print("Slope (coefficient): ",model.coef_[0])
```

```python

#after running model.fir - model calculates intercept
print("Intercept: ",model.intercept_)
```
